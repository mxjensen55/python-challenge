```python
#import depencies
import os
import csv

#set path for the csv file

election_data=os.path.join("Resources", "election_data.csv")

#set variables

total_votes=0
khan_votes=0
correy_votes=0
li_votes=0
otooley_votes=0

#read file

with open (election_data) as csvfile:
    csvreader=csv.reader(csvfile, delimiter= ",")
    
    #remove header
    
    csv_header = next(csvfile, None)
    
    #loop over rows to find totals
    
    for row in csvreader:
        
        total_votes = total_votes+1
        
        #if, elif statment to find votes for particular candidate
        
        if row[2] == "Khan":
            khan_votes=khan_votes+1
        elif row[2] == "Correy":
            correy_votes=correy_votes+1
        elif row[2] == "Li":
            li_votes=li_votes+1
        elif row[2]=="O'Tooley":
                otooley_votes=otooley_votes+1

#create dictionaries
                
candidates=["Khan", "Correy", "Li", "O'Tooley"]
vote_count=[khan_votes, correy_votes, li_votes, otooley_votes]

#zip together dictionaries and use max function to find candidate with the highest number of votes--set this to a variable.

comb_cand_votes=dict(zip(candidates,vote_count))
key= max(comb_cand_votes, key=comb_cand_votes.get)

#create variables to use in summary analysis

khan_results=(khan_votes/ total_votes)*100
correy_results=(correy_votes/total_votes)*100
li_results=(li_votes/total_votes)*100
otooley_results=(otooley_votes/total_votes)*100

#print analysis

print("Election Results")
print("------------------")
print(f"Total Votes: {total_votes}")
print("------------------")
print(f"Khan: {khan_results: .3f}% ({khan_votes})")
print(f"Correy: {correy_results: .3f}% ({correy_votes})")
print(f"Li: {li_results: .3f}% ({li_votes})")
print(f"O'Tooley: {otooley_results: .3f}% ({otooley_votes})")
print("-------------------")
print(f"Winner: {key}")
print("-------------------")
       
    
```

```python
#create text file
output_file = os.path.join("election_analysis.txt")
#open output file
with open(output_file, "w") as datafile:
    writer=csv.writer(datafile)
    #write results to txt file
    datafile.write("Election Results")
    datafile.write("\n")
    datafile.write("-----------------")
    datafile.write("\n")
    datafile.write(f"Total Votes: {total_votes}")
    datafile.write("\n")
    datafile.write("-----------------")
    datafile.write("\n")
    datafile.write(f"Khan: {khan_results: .3f}% ({khan_votes})")
    datafile.write("\n")
    datafile.write(f"Correy: {correy_results: .3f}% ({correy_votes})")
    datafile.write("\n")
    datafile.write(f"Li: {li_results: .3f}% ({li_votes})")
    datafile.write("\n")
    datafile.write(f"O'Tooley: {otooley_results: .3f}% ({otooley_votes})")
    datafile.write("\n")
    datafile.write("------------------")
    datafile.write("\n")
    datafile.write(f"Winner: {key}")
    datafile.write("\n")
    datafile.write("------------------")
    
```
